## Now to the Action 

```python
def get_data(total_dis_light,total_dis_med):
    """
    This function fed on the row name in the page that I want to scrap.
    I believe there is a better/more elegant way to get the data, rather than using code below, 
    with a little more time to think
    """
    for i in range(len(total_dis_light)):
        Serial.append(total_dis_light[i].findAll('td')[0].text)
        Event.append(total_dis_light[i].findAll('td')[1].text)
        Province.append(total_dis_light[i].findAll('td')[2].text)
        District.append(total_dis_light[i].findAll('td')[3].text)
        SubDistrict.append(total_dis_light[i].findAll('td')[4].text)
        Date.append(total_dis_light[i].findAll('td')[5].text)
        Location.append(total_dis_light[i].findAll('td')[6].text)
        Deaths.append(total_dis_light[i].findAll('td')[7].text)
        Injured.append(total_dis_light[i].findAll('td')[8].text)
        Missing.append(total_dis_light[i].findAll('td')[9].text)
        House_Destroyed.append(total_dis_light[i].findAll('td')[10].text)
        House_Damaged.append(total_dis_light[i].findAll('td')[11].text)
        Directly_Affected.append(total_dis_light[i].findAll('td')[12].text)
        Indirectly_Affected.append(total_dis_light[i].findAll('td')[13].text)
        Relocated.append(total_dis_light[i].findAll('td')[14].text)
        Evacuated.append(total_dis_light[i].findAll('td')[15].text)
        LossesUSD.append(total_dis_light[i].findAll('td')[16].text)
        LossesLocal.append(total_dis_light[i].findAll('td')[17].text)
        EducationCenter.append(total_dis_light[i].findAll('td')[18].text)
        Hospitals.append(total_dis_light[i].findAll('td')[19].text)
        CropsDamageHa.append(total_dis_light[i].findAll('td')[20].text)
        CattleLoss.append(total_dis_light[i].findAll('td')[21].text)
        RoadDamageM.append(total_dis_light[i].findAll('td')[22].text)
        GLIDENum.append(total_dis_light[i].findAll('td')[23].text)
        Comments.append(total_dis_light[i].findAll('td')[24].text)
        
    for i in range(len(total_dis_med)):
        Serial.append(total_dis_med[i].findAll('td')[0].text)
        Event.append(total_dis_med[i].findAll('td')[1].text)
        Province.append(total_dis_med[i].findAll('td')[2].text)
        District.append(total_dis_med[i].findAll('td')[3].text)
        SubDistrict.append(total_dis_med[i].findAll('td')[4].text)
        Date.append(total_dis_med[i].findAll('td')[5].text)
        Location.append(total_dis_med[i].findAll('td')[6].text)
        Deaths.append(total_dis_med[i].findAll('td')[7].text)
        Injured.append(total_dis_med[i].findAll('td')[8].text)
        Missing.append(total_dis_med[i].findAll('td')[9].text)
        House_Destroyed.append(total_dis_med[i].findAll('td')[10].text)
        House_Damaged.append(total_dis_med[i].findAll('td')[11].text)
        Directly_Affected.append(total_dis_med[i].findAll('td')[12].text)
        Indirectly_Affected.append(total_dis_med[i].findAll('td')[13].text)
        Relocated.append(total_dis_med[i].findAll('td')[14].text)
        Evacuated.append(total_dis_med[i].findAll('td')[15].text)
        LossesUSD.append(total_dis_med[i].findAll('td')[16].text)
        LossesLocal.append(total_dis_med[i].findAll('td')[17].text)
        EducationCenter.append(total_dis_med[i].findAll('td')[18].text)
        Hospitals.append(total_dis_med[i].findAll('td')[19].text)
        CropsDamageHa.append(total_dis_med[i].findAll('td')[20].text)
        CattleLoss.append(total_dis_med[i].findAll('td')[21].text)
        RoadDamageM.append(total_dis_med[i].findAll('td')[22].text)
        GLIDENum.append(total_dis_med[i].findAll('td')[23].text)
        Comments.append(total_dis_med[i].findAll('td')[24].text)
```

```python
# Declare the number of pages that available, in this case its 331
pages = 331

#create empty lists 
Serial = []
Event = []
Province = []
District = []
SubDistrict = []
Date = []
Location = []
Deaths = []
Injured = []
Missing = []
House_Destroyed = []
House_Damaged = []
Directly_Affected = []
Indirectly_Affected = []
Relocated = []
Evacuated = []
LossesUSD = []
LossesLocal = []
EducationCenter = []
Hospitals = []
CropsDamageHa = []
CattleLoss = []
RoadDamageM = []
GLIDENum = []
Comments = []

#Get the first page
driver = webdriver.Chrome('./chromedriver')  # Optional argument, if not specified will search path.
driver.get('https://www.desinventar.net/DesInventar/profiletab.jsp?countrycode=idn&continue=y')

#Get the first page of the data panel
driver.find_element_by_link_text("View data").click()

#Get the html elements and all parametres
url = driver.page_source
# request = url.text
total_soup = Soup(url,'html.parser')
total_dis_light = total_soup.findAll('tr', {'class':'bodylight'})
total_dis_med = total_soup.findAll('tr', {'class':'bodymedlight'})

#Get the data
print("getting the 1st page")
get_data(total_dis_light,total_dis_med)

#Now Get the data from page 2 to 331
for page in range(1, pages):
    if page < (pages-1):
        #This Script click on the Next page button
        driver.find_element_by_xpath("//img[@src='/DesInventar/images/arrow-forward.gif']").click()

        #redetermine the html elements
        url = driver.page_source
        total_soup = Soup(url,'html.parser')
        total_dis_light = total_soup.findAll('tr', {'class':'bodylight'})
        total_dis_med = total_soup.findAll('tr', {'class':'bodymedlight'})

        print("Now, gettting the page {}".format(page+1))
        #Aaaand get the data
        get_data(total_dis_light,total_dis_med)
    else:
         #This Script is to access the final page 
        driver.find_element_by_xpath("//img[@src='/DesInventar/images/arrow-last.gif']").click()

        #redetermine the html elements
        url = driver.page_source
        total_soup = Soup(url,'html.parser')
        total_dis_light = total_soup.findAll('tr', {'class':'bodylight'})
        total_dis_med = total_soup.findAll('tr', {'class':'bodymedlight'})

        print("Now, gettting the final page")
        #Aaaand get the data
        get_data(total_dis_light,total_dis_med)
```

```python
Params = [Serial
          ,Event
          ,Province
          ,District
          ,SubDistrict
          ,Date
          ,Location
          ,Deaths
          ,Injured
          ,Missing
          ,House_Destroyed
          ,House_Damaged
          ,Directly_Affected
          ,Indirectly_Affected
          ,Relocated
          ,Evacuated
          ,LossesUSD
          ,LossesLocal
          ,EducationCenter
          ,Hospitals
          ,CropsDamageHa
          ,CattleLoss
          ,RoadDamageM
          ,GLIDENum
          ,Comments]

Params_name = ["serial"
          ,"event"
          ,"province"
          ,"district"
          ,"sub_district"
          ,"date"
          ,"location"
          ,"deaths"
          ,"injured"
          ,"missing"
          ,"house_destroyed"
          ,"house_damaged"
          ,"directly_affected"
          ,"indirectly_affected"
          ,"relocated"
          ,"evacuated"
          ,"losses_USD"
          ,"losses_local"
          ,"education_center"
          ,"hospitals"
          ,"crops_damageHa"
          ,"cattle_loss"
          ,"road_damage_M"
          ,"GLIDE_num"
          ,"Comments"]

df = pd.DataFrame(Params).T
df.columns = Params_name
```

```python
df.head(1)
```

```python
#convert the date to date type
df["date"] = pd.to_datetime(df["date"])
```

```python
#define categorical and numerical columns
categorical = ['event','province']
numerical = ['deaths', 'injured', 'missing', 'house_destroyed', 'house_damaged'
             , 'directly_affected', 'indirectly_affected','evacuated', 'losses_local'
             , 'education_center', 'hospitals', 'crops_damageHa', 'road_damage_M']
```

```python
#convert the categorical and numerical columns to what they really are
for category in categorical:
    df[category] = df[category].astype('category')

df[numerical] = df[numerical].apply(pd.to_numeric, errors='coerce').fillna(0)
```

```python
# delete the location, sub_district, losses USD, cattle Loss, GLIDE Num, comments and relocated

disaster_data = df[['date'
                    ,'event'
                    ,'province'
                    ,'district'
                    ,'deaths'
                    ,'injured'
                    ,'missing'
                    ,'house_destroyed'
                    ,'house_damaged'
                    ,'directly_affected'
                    ,'indirectly_affected'
                    ,'relocated'
                    ,'evacuated'
                    ,'losses_local'
                    ,'education_center'
                    ,'hospitals'
                    ,'crops_damageHa'
                    ,'road_damage_M']].sort_values(by='date', ascending = False).reset_index(drop=True)
```
